```python
# 引入必要模組
from collections import defaultdict
import matplotlib.pyplot as plt

# 設定要讀取的檔案名稱
filename = "hw2_data.txt"

# 建立一個字典來統計每個字出現的次數（hash）
word_count = defaultdict(int)

# 開啟檔案，逐行讀取
with open(filename, 'r', encoding='utf-8') as f:
    for line in f:
        word = line.strip()  # 去除換行符號與空白
        word_count[word] += 1  # 累加該字的出現次數

# 將字典依照出現次數由大到小排序
sorted_word_count = dict(sorted(word_count.items(), key=lambda item: item[1], reverse=True))

# 顯示結果：不重複的字數量
print("總共有", len(sorted_word_count), "個不重複的字\n")

# 顯示每一個字與它的出現次數
print("每一個英文字出現次數為：")
for word, count in sorted_word_count.items():
    print(f"{word}: {count}")

# 畫出直方圖（bar chart）
plt.figure(figsize=(12, 6))  # 設定圖表大小
plt.bar(sorted_word_count.keys(), sorted_word_count.values())  # 繪製長條圖
plt.xlabel('Words')  # x軸標籤
plt.ylabel('Frequency')  # y軸標籤
plt.title('Word Frequency Histogram by Min-Hsuan,Ku 20250406')  # 圖表標題
plt.xticks(rotation=90)  # 旋轉 x 軸文字避免重疊
plt.tight_layout()  # 自動調整版面配置
plt.show()  # 顯示圖表
```
